## ğŸ”§ ì‹¤ìŠµ ë¬¸ì œ

### ì‹¤ìŠµ 1: ê°„ë‹¨í•œ ë©”ëª¨ í”„ë¡œê·¸ë¨

**ë¬¸ì œ**: ë©”ëª¨ë¥¼ íŒŒì¼ì— ì €ì¥í•˜ê³  ì½ì–´ì˜¤ëŠ” ê°„ë‹¨í•œ í”„ë¡œê·¸ë¨ì„ ë§Œë“œì„¸ìš”.

**ì •ë‹µ**:

````python
def write_memo(memo):
    """ë©”ëª¨ë¥¼ íŒŒì¼ì— ì €ì¥"""
    with open("memo.txt", "w", encoding="utf-8") as file:
        file.write(memo)
    print("ë©”ëª¨ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤!")

def read_memo():
    """ë©”ëª¨ íŒŒì¼ ì½ê¸°"""
    try:
        with open("memo.txt", "r", encoding="utf-8") as file:
            content = file.read()
        print("ì €ì¥ëœ ë©”ëª¨:")
        print(content)
    except FileNotFoundError:
        print("ì €ì¥ëœ ë©”ëª¨ê°€ ì—†ìŠµë‹ˆë‹¤.")

def add_memo(new_memo):
    """ê¸°ì¡´ ë©”ëª¨ì— ì¶”ê°€"""
    with open("memo.txt", "a", encoding="utf-8") as file:
        file.write("\n" + new_memo)
    print("ë©”ëª¨ê°€ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤!")

# í”„ë¡œê·¸ë¨ í…ŒìŠ¤íŠ¸
write_memo("ì˜¤ëŠ˜ íŒŒì´ì¬ íŒŒì¼ ì²˜ë¦¬ë¥¼ ë°°ì› ë‹¤.")
read_memo()
add_memo("ë‚´ì¼ì€ ë” ì—°ìŠµí•´ë³´ì!")
read_memo()
````

### ì‹¤ìŠµ 2: ì‡¼í•‘ ëª©ë¡ ê´€ë¦¬

**ë¬¸ì œ**: ì‡¼í•‘ ëª©ë¡ì„ íŒŒì¼ì— ì €ì¥í•˜ê³  ê´€ë¦¬í•˜ëŠ” í”„ë¡œê·¸ë¨ì„ ë§Œë“œì„¸ìš”.

**ì •ë‹µ**:

````python
def add_item(item):
    """ì‡¼í•‘ ëª©ë¡ì— í•­ëª© ì¶”ê°€"""
    with open("shopping_list.txt", "a", encoding="utf-8") as file:
        file.write(item + "\n")
    print(f"'{item}'ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤.")

def show_list():
    """ì‡¼í•‘ ëª©ë¡ ë³´ê¸°"""
    try:
        with open("shopping_list.txt", "r", encoding="utf-8") as file:
            items = file.readlines()
      
        print("=== ì‡¼í•‘ ëª©ë¡ ===")
        for i, item in enumerate(items, 1):
            print(f"{i}. {item.strip()}")
    except FileNotFoundError:
        print("ì‡¼í•‘ ëª©ë¡ì´ ë¹„ì–´ìˆìŠµë‹ˆë‹¤.")

def count_items():
    """í•­ëª© ê°œìˆ˜ ì„¸ê¸°"""
    try:
        with open("shopping_list.txt", "r", encoding="utf-8") as file:
            items = file.readlines()
        print(f"ì´ {len(items)}ê°œì˜ í•­ëª©ì´ ìˆìŠµë‹ˆë‹¤.")
    except FileNotFoundError:
        print("ì‡¼í•‘ ëª©ë¡ì´ ë¹„ì–´ìˆìŠµë‹ˆë‹¤.")

# í”„ë¡œê·¸ë¨ í…ŒìŠ¤íŠ¸
add_item("ì‚¬ê³¼")
add_item("ë°”ë‚˜ë‚˜")
add_item("ìš°ìœ ")
show_list()
count_items()
````

### ì‹¤ìŠµ 3: ì ìˆ˜ ê¸°ë¡ í”„ë¡œê·¸ë¨

**ë¬¸ì œ**: ê²Œì„ ì ìˆ˜ë¥¼ íŒŒì¼ì— ê¸°ë¡í•˜ê³  ìµœê³  ì ìˆ˜ë¥¼ ì°¾ëŠ” í”„ë¡œê·¸ë¨ì„ ë§Œë“œì„¸ìš”.

**ì •ë‹µ**:

````python
def save_score(score):
    """ì ìˆ˜ë¥¼ íŒŒì¼ì— ì €ì¥"""
    with open("scores.txt", "a", encoding="utf-8") as file:
        file.write(str(score) + "\n")
    print(f"ì ìˆ˜ {score}ê°€ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

def show_all_scores():
    """ëª¨ë“  ì ìˆ˜ ë³´ê¸°"""
    try:
        with open("scores.txt", "r", encoding="utf-8") as file:
            scores = file.readlines()
      
        print("=== ì ìˆ˜ ê¸°ë¡ ===")
        for i, score in enumerate(scores, 1):
            print(f"ê²Œì„ {i}: {score.strip()}ì ")
    except FileNotFoundError:
        print("ì ìˆ˜ ê¸°ë¡ì´ ì—†ìŠµë‹ˆë‹¤.")

def find_best_score():
    """ìµœê³  ì ìˆ˜ ì°¾ê¸°"""
    try:
        with open("scores.txt", "r", encoding="utf-8") as file:
            scores = file.readlines()
      
        # ë¬¸ìì—´ì„ ìˆ«ìë¡œ ë³€í™˜
        score_numbers = []
        for score in scores:
            score_numbers.append(int(score.strip()))
      
        best_score = max(score_numbers)
        print(f"ìµœê³  ì ìˆ˜: {best_score}ì ")
    except FileNotFoundError:
        print("ì ìˆ˜ ê¸°ë¡ì´ ì—†ìŠµë‹ˆë‹¤.")
    except ValueError:
        print("ì ìˆ˜ íŒŒì¼ì— ì˜¤ë¥˜ê°€ ìˆìŠµë‹ˆë‹¤.")

# í”„ë¡œê·¸ë¨ í…ŒìŠ¤íŠ¸
save_score(85)
save_score(92)
save_score(78)
show_all_scores()
find_best_score()
````
